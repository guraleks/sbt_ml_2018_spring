```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import RandomizedSearchCV
import numpy as np
import pandas as pd
```

# Подготовка данных

```python
data = pd.read_csv("bioresponse.csv")
```

```python
data.head(7)
```

```python
y = data.Activity.values
```

```python
print(y)
```

```python
X = data.iloc[:, 1:]
```

```python
print(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_test
```

```python
print(len(y_test) / len(y))
```

# Работа с моделями

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
```

### 1) K neighbors classifier

```python
model = KNeighborsClassifier()
```

```python
params = list(model.get_params().keys())
```

```python
params
```

```python
params_grid = {'algorithm' : ['auto', 'ball_tree', 'kd_tree', 'brute'],
              'n_neighbors' : [1, 2, 5, 10, 15],
              }
```

```python
grid_cv = GridSearchCV(model, params_grid)
```

```python
%%timeit
grid_cv.fit(X_train, y_train)
```

```python
print(grid_cv.best_estimator_)
print(grid_cv.best_score_)
print(grid_cv.best_params_)
```

```python
rand_grid_cv = RandomizedSearchCV(model, params_grid)
```

```python
%%time
rand_grid_cv.fit(X_train, y_train)
```

```python
print(rand_grid_cv.best_estimator_)
print(rand_grid_cv.best_score_)
print(rand_grid_cv.best_params_)
```

### 2) Random forest classifier

```python
model = RandomForestClassifier()
```

```python
params = list(model.get_params().keys())
```

```python
print(params)
```

```python
params_grid = {
    'n_estimators' : [30, 31, 32],
    'bootstrap' : [False, True],
    'criterion' : ['gini', 'entropy'],
    'n_jobs' : [1, 2, 3],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
grid_cv = GridSearchCV(model, params_grid)
```

```python
%%time
grid_cv.fit(X_train, y_train)
```

```python
print(grid_cv.best_estimator_)
print(grid_cv.best_params_)
print(grid_cv.best_score_)
```

```python
rand_grid_cv = RandomizedSearchCV(model, params_grid)
```

```python
%%time
rand_grid_cv.fit(X_train, y_train)
```

```python
print(rand_grid_cv.best_estimator_)
print(rand_grid_cv.best_params_)
print(rand_grid_cv.best_score_)
```

### 3) Decision Tree Classifier

```python
model = DecisionTreeClassifier()
```

```python
params = list(model.get_params().keys())
```

```python
print(params)
```

```python
params_grid = {
    'max_depth' : [1, 2, 5, 10, 15, 20],
    'min_samples_split' : [2, 3, 4, 5, 10, 15],
    'criterion' : ['gini', 'entropy'],
    'max_features' : ['auto', 'sqrt', 'log2']
}
```

```python
grid_cv = GridSearchCV(model, params_grid)
```

```python
%%time
grid_cv.fit(X_train, y_train)
```

```python
print(grid_cv.best_estimator_)
print(grid_cv.best_params_)
print(grid_cv.best_score_)
```

```python
rand_grid_cv = RandomizedSearchCV(model, params_grid)
```

```python
%%time
rand_grid_cv.fit(X_train, y_train)
```

```python
print(rand_grid_cv.best_estimator_)
print(rand_grid_cv.best_params_)
print(rand_grid_cv.best_score_)
```

# Выводы:

GridSearchCV делает полный перебор по всей сетке параметров, поэтому нежелательно ее использование на больших объемах данных и большом количестве параметров (большом размере сетки)

RandomizedSearchCV, напротив, лучше использовать при большом объеме данных и сетки параметров, так как в этом методе происходит случайный перебор по сетке параметров

В то же время, что естественно, качество модели при использовании GridSearchCV оказывается лучше
